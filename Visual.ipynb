```python
import random
import matplotlib
import math
import matplotlib.pyplot as plt
xlst = []
ylst = []
for i in range(0,10000):
  x = random.random()
  y = random.random()
  xlst.append(x)
  ylst.append(y)
```

```python
figure, axes = plt.subplots()
plt.scatter(xlst,ylst,marker= "o",s=0.2)
axes.set_aspect(1)
plt.xlim(0,1)
plt.ylim(0,1)
c1 = plt.Circle((0.5,0.5),0.2,color='r',fill=False)
c2 = plt.Circle((0.5,0.5),0.4,color='g',fill=False)
axes.add_artist(c1)
axes.add_artist(c2)
plt.title("Distribution of sample")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["radius=0.2,0.4"], loc ="upper right")
plt.show
```
