```python
# import the required libraries 
import random 
import matplotlib.pyplot as plt 
    
# store the random numbers in a  
# list 
xlst = []
ylst=[]
mu = 1
sigma = 0.1
    
for i in range(10000): 
    x= random.gauss(mu, sigma)
    y= random.gauss(mu, sigma)
    xlst.append(x)
    ylst.append(y)   
# plotting a graph 
figure, axes = plt.subplots()
plt.plot(xlst,ylst,'o',markersize=0.2)
axes.set_aspect(1)
plt.xlim(0.6,1.4)
plt.ylim(0.6,1.4)
c1 = plt.Circle((1,1),0.08,color='r',fill=False)
c2 = plt.Circle((1,1),0.23,color='g',fill=False)
axes.add_artist(c1)
axes.add_artist(c2)
plt.title("Distribution of sample")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["radius=0.08,0.23"], loc ="upper right")

plt.show()
```
