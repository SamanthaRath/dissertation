```python
from astropy.io import fits
hdul = fits.open('dr14q_spec_prop.fits')
```

```python
data = hdul[1].data
modidataR = data['REDSHIFT']
modidataN=data['SDSS_NAME']
modidataRA = data['RA']
modidataDEC = data['DEC']
```

```python
redshift=[]
ra=[]
dec=[]
for i in range(526265):
  if(modidataR[i]<1):
    redshift.append(modidataR[i])
    ra.append(modidataRA[i])
    dec.append(modidataDEC[i])
```

```python
print(len(redshift))
print(redshift)
```

```python
import math
zr1=[]
zr2=[]
n=5000
for i in range(n-1):
  for j in range(i+1,n):
    dr=abs(redshift[j]-redshift[i])
    if(dr<0.02):
      zr1.append(redshift[i])
      zr2.append(redshift[j])
      print("Quasar 1 is: ",redshift[i],"Quasar 2 is: ",redshift[j])
      a1=math.sin(math.radians(dec[i]))
      a2=math.cos(math.radians(dec[i]))
      b1=math.sin(math.radians(dec[j]))
      b2=math.cos(math.radians(dec[j]))
      c=math.cos(math.radians(abs(ra[i]-ra[j])))
      x=(a1*b1)+(a2*b2*c)
      #print("Value of X ", x)
      d=math.degrees(math.acos(x))
      min=d*60
      print("Angular Separation = ",d, '\n')
      print("------------------------------------")
      if(min<10):
        print(" This is a pair")
        print("------------------------------------")
print(zr1)
print(zr2)
```
